```python
# ----- 環境変数読み込み -----
from dotenv import load_dotenv
load_dotenv()
```

```python
# ----- HTMLを読み込む
from langchain.document_loaders import UnstructuredURLLoader
loader = UnstructuredURLLoader(urls = ['https://github.com/spring-projects/spring-boot/wiki/Spring-Boot-3.0-Migration-Guide'])
data = loader.load_and_split()
data
```

```python
# URLからWebコンテンツを読み込み、コンテンツをもとに回答させる
from langchain.indexes import VectorstoreIndexCreator
index = VectorstoreIndexCreator(vectorstore_kwargs = {"persist_directory": ".data"}).from_documents(data)

query = "What should be done to migrate to Spring Boot 3.0?"
index.query_with_sources(query)
```
